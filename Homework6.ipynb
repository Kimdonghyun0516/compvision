```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('dsu1.jpg',1)
rows,cols,ch = img.shape
pts1 = np.float32([[318,276],[759,313],[765,508],[332,537]])
pts2 = np.float32([[1,1],[443,2],[443,193],[2,193]])
M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(443,193))
```

```python
plt.figure(figsize=(25,20))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

```python
cv2.imshow('img',img)
cv2.imshow('img',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
