## 1.4.3 Geo

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('test.jpg')
resN = cv2.resize(img,None,fx=2, fy=2, interpolation = cv2.INTER_NEAREST)
resC = cv2.resize(img,None,fx=2, fy=2, interpolation = cv2.INTER_CUBIC)
```

```python
cv2.imshow('image',img)
cv2.imshow('resN image',resN)
cv2.imshow('resC image',resC)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#OR
height, width = img.shape[:2]
res = cv2.resize(img,(2*width, 2*height), interpolation = cv2.INTER_CUBIC)
```

```python
cv2.imshow('image',img)
cv2.imshow('res image',res)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Translation

```python
img = cv2.imread('test.jpg',1)
tx = 100
ty = 150
rows,cols,ch = img.shape
M = np.float32([[1,0,100],[0,1,150]])
dst = cv2.warpAffine(img,M,(cols,rows))
dst=cv2.rectangle(dst,(0,0),(tx,ty),(0,255,0),-1)
cv2.imshow('img',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Rotation

```python
img = cv2.imread('test.jpg',1)
rows,cols,ch = img.shape
cx = 100
cy = 100
theta = 45
M = cv2.getRotationMatrix2D((cx,cy),theta,1) 
dst = cv2.warpAffine(img,M,(cols,rows))
dst = cv2.circle(dst,(cx,cy), 10, (0,0,255), -1)
cv2.imshow('img',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Affine Transformation 중요하다고 하심

```python
img = cv2.imread('dsu4.jpg')
rows,cols,ch = img.shape
pts1 = np.float32([[235,457],[238,269],[569,231]])
pts2 = np.float32([[235,457],[235,270],[570,270]])
M = cv2.getAffineTransform(pts1,pts2)
dst = cv2.warpAffine(img,M,(cols,rows))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

```python
cv2.imshow('simg',img)
cv2.imshow('img',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Perspective Transformation

```python
img = cv2.imread('dsu5.jpg')
rows,cols,ch = img.shape
pts1 = np.float32([[579,338],[753,375],[763,543],[584,543]])
pts2 = np.float32([[579,340],[760,340],[760,540],[579,540]])
M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(0,0))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

```python
cv2.imshow('simg',img)
cv2.imshow('img',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
